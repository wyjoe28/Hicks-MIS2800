```python
import sqlite3
```

```python
conn = sqlite3.connect('AFC_North_APPG.db')
```

```python
conn.execute('''
CREATE TABLE Hicks_HW_AFC_North_APPG(
  team_name TEXT PRIMARY KEY NOT NULL,
  coach TEXT,
  ties INTEGER,
  wins INTEGER,
  losses INTEGER)
  STRICT;
  ''')
```

```python
conn.execute('''
INSERT INTO Hicks_HW_AFC_North_APPG (team_name, coach, wins, losses, ties)
VALUES ('Steelers', 'Mike Tomlin', 10, 7, 0),
('Ravens', 'John Harbaugh', 8, 9, 0),
('Bengals', 'Zac Taylor', 6, 11, 0),
('Browns', 'Kevin Stefanski', 5, 12, 0);
''')
```

```python
query = "SELECT team_name, coach, wins, losses, ties FROM Hicks_HW_AFC_North_APPG WHERE wins > 7"
results = conn.execute(query)
for row in results:
  print(row)
```

```python
query = "SELECT team_name, wins FROM Hicks_HW_AFC_North_APPG ORDER BY wins DESC"
for row in conn.execute(query):
  print(row)
```
